```python
# -*- coding: utf-8 -*-
#Import Necessary Libraries
import numpy as np
import pandas as pd

#visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#ignore warnings
import warnings
warnings.filterwarnings('ignore')


#Read in and Explore the Data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
features_details = train.describe(include = "all");
nan_value_detail = pd.isnull(train).sum();
```

```python
print(nan_value_detail)
```

```python
mean = np.mean(train['weight'], axis=0)
sd = np.std(train['weight'], axis=0)
cut_off = sd * 3
lower, upper = mean - cut_off, mean + cut_off
weight_outliers = [x for x in train['weight'] if x < lower or x > upper]
print(len(weight_outliers))

for i in weight_outliers:
    train.drop(train[train.weight == i].index, inplace=True)
    test.drop(test[test.weight == i].index, inplace=True)
    
features_details = train.describe(include = "all");
print(features_details)
```

```python
hmean = np.mean(train['height'], axis=0)
hsd = np.std(train['height'], axis=0)
cut_off = hsd * 3
lower, upper = hmean - cut_off, hmean + cut_off
height_outliers = [x for x in train['height'] if x < lower or x > upper]
for i in height_outliers:
    train.drop(train[train.height == i].index, inplace=True)
    test.drop(test[test.height == i].index, inplace=True)
features_details = train.describe(include = "all");
print(features_details)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict
x = train.drop(['id', 'weight'], axis=1)
y = train["weight"]
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
# Fitting Polynomial Regression to the dataset
'''from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)'''
```

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=50)
regressor.fit(x_train, y_train)
```

```python
# Visualising the Polynomial Regression results
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
y_pred = cross_val_predict(regressor, x_val, y_val, cv=10)
error = mean_squared_error(y_val, y_pred)
fig, ax = plt.subplots()
ax.scatter(y_val, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_pred.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
error = mean_squared_error(y_val, y_pred)
print("MEAN Squared Error : {}. (Lower the better)".format(error))
```

```python
#test prediction
x_test = test.drop(['id', 'weight'], axis=1)
y_test = test["weight"]
y_test_pred = cross_val_predict(regressor, x_test, y_test, cv=10)
error = mean_squared_error(y_test, y_test_pred)
fig, ax = plt.subplots()
ax.scatter(y_test, y_test_pred, edgecolors=(0, 0, 0))
ax.plot([y_test_pred.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
print("MEAN Squared Error : {}. (Lower the better)".format(error))
```
